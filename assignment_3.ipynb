```python
def mean(data):
  """Calculates the mean of a list of numbers."""
  sum = 0
  count = len(data)
  for number in data:
    sum += number
  return sum / count

region_a = [10, 15, 12, 8, 14]
region_b = [18, 20, 16, 22, 25]

mean_region_a = mean(region_a)
mean_region_b = mean(region_b)

print("The mean sales for region A is", mean_region_a)
print("The mean sales for region B is", mean_region_b)

```

```python
def mode(data):
  """Calculates the mode of a list of numbers."""
  counts = {}
  for number in data:
    if number in counts:
      counts[number] += 1
    else:
      counts[number] = 1

  max_count = 0
  mode = None
  for number, count in counts.items():
    if count > max_count:
      max_count = count
      mode = number

  return mode

data = [4, 5, 2, 3, 5, 4, 3, 2, 4, 5]

mode_of_survey_responses = mode(data)

print("The mode of the survey responses is", mode_of_survey_responses)
```

```python
def median(data):
  """Calculates the median of a list of numbers."""
  data.sort()
  length = len(data)
  if length % 2 == 0:
    # The list has an even number of elements, so the median is the average of the two middle elements.
    median = (data[length // 2] + data[length // 2 - 1]) / 2
  else:
    # The list has an odd number of elements, so the median is the middle element.
    median = data[length // 2]
  return median

department_a = [5000, 6000, 5500, 7000]
department_b = [4500, 5500, 5800, 6000, 5200]

median_department_a = median(department_a)
median_department_b = median(department_b)

print("The median salary for department A is", median_department_a)
print("The median salary for department B is", median_department_b)
```

```python
import math

def t_test(group_a, group_b):
  """Performs a t-test to compare the means of two groups."""
  mean_a = sum(group_a) / len(group_a)
  mean_b = sum(group_b) / len(group_b)
  std_dev_a = math.sqrt(sum((x - mean_a)**2 for x in group_a)) / len(group_a)
  std_dev_b = math.sqrt(sum((x - mean_b)**2 for x in group_b)) / len(group_b)
  t_statistic = (mean_a - mean_b) / (std_dev_a**2 + std_dev_b**2)**0.5
  p_value = 1 - math.erf(t_statistic / math.sqrt(2))
  return t_statistic, p_value

group_a = [85, 90, 92, 88, 91]
group_b = [82, 88, 90, 86, 87]

t_statistic, p_value = t_test(group_a, group_b)

print("The t-statistic is", t_statistic)
print("The p-value is", p_value)
```

```python
import math

def correlation_coefficient(x_data, y_data):
  """Calculates the correlation coefficient between two sets of data."""
  mean_x = sum(x_data) / len(x_data)
  mean_y = sum(y_data) / len(y_data)
  std_dev_x = math.sqrt(sum((x - mean_x)**2 for x in x_data)) / len(x_data)
  std_dev_y = math.sqrt(sum((y - mean_y)**2 for y in y_data)) / len(y_data)
  covariance = sum((x - mean_x)*(y - mean_y) for x, y in zip(x_data, y_data))
  correlation_coefficient = covariance / (std_dev_x * std_dev_y)
  return correlation_coefficient

advertising_expenditure = [10, 15, 12, 8, 14]
sales = [25, 30, 28, 20, 26]

correlation_coefficient = correlation_coefficient(advertising_expenditure, sales)

print("The correlation coefficient is", correlation_coefficient)
```

```python
import math

def standard_deviation(data):
  """Calculates the standard deviation of a list of numbers."""
  mean = sum(data) / len(data)
  squared_deviations = []
  for number in data:
    squared_deviation = (number - mean)**2
    squared_deviations.append(squared_deviation)
  variance = sum(squared_deviations) / len(data)
  standard_deviation = math.sqrt(variance)
  return standard_deviation

heights = [160, 170, 165, 155, 175, 180, 170]

standard_deviation = standard_deviation(heights)

print("The standard deviation is", standard_deviation)
```
